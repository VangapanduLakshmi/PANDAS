### STEP 1 : IDENTIFY INPUT AND OUTPUT VARIABLES  

```python
x = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'good', 'color']]
y = df['quality']
```

### STEP 2 : SPLITING THE DATA INTO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

### STEP 3 : DATA PREPARATION FOR X_TRAIN

```python
X_train_cat = X_train.select_dtypes(include=['object'])
X_train_num = X_train.select_dtypes(include=['int64', 'float64'])
```

```python
X_train_cat.head()
```

```python
X_train_num.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder_ = OneHotEncoder(drop='first', sparse_output=False)

# column names are (annoyingly) lost after OneHotEncoding
# (i.e. the dataframe is converted to a numpy ndarray)
X_train_cat_tansformed = pd.DataFrame(encoder_.fit_transform(X_train_cat), 
                               columns=encoder_.get_feature_names_out(), 
                               index = X_train_cat.index)
X_train_cat_tansformed.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# column names are (annoyingly) lost after Scaling
# (i.e. the dataframe is converted to a numpy ndarray)
X_train_num_transformed = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = scaler.get_feature_names_out(), 
                                    index = X_train_num.index)

X_train_num_transformed.head()
```

```python
X_train_transformed = pd.concat([X_train_num_transformed, X_train_cat_tansformed], axis=1)

X_train_transformed.head()
```

### STEP 4: DATA PREPARATION FOR X_TEST

```python
X_test_cat = X_test.select_dtypes(include=['object'])
X_test_num = X_test.select_dtypes(include=['int64', 'float64'])
```

```python
X_test_num_transformed = pd.DataFrame(scaler.transform(X_test_num), 
                                   columns = scaler.get_feature_names_out(), 
                                   index = X_test_num.index)

X_test_num_transformed.head()
```

```python
X_test_cat_tansformed = pd.DataFrame(encoder_.transform(X_test_cat), 
                               columns=encoder_.get_feature_names_out(), 
                               index = X_test_cat.index)
X_test_cat_tansformed.head()
```

```python
X_test_transformed = pd.concat([X_test_num_transformed, X_test_cat_tansformed], axis=1)

X_test_transformed.head()
```

### HYPERPRAMETER TUNING THROUGH GRIDSEARCH CV

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

### IN KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
tuned_parameters = [{'n_neighbors': [i for i in range(1, 51)], 'p': [1, 2, 3]}]

clf = GridSearchCV(
    estimator=KNeighborsRegressor(),
    param_grid=tuned_parameters,
    scoring='neg_mean_squared_error',  # Use appropriate regression scoring metric
    cv=5,
    return_train_score=True,
    verbose=1
)

clf.fit(X_train_transformed, y_train)
```

```python
# Get the best parameters and best model
best_params = clf.best_params_
best_model = clf.best_estimator_

# Make predictions
predictions = best_model.predict(X_test_transformed)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Best Parameters:", best_params)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

```python
output_df = pd.DataFrame({'Actual': y_test})
output_df['KNN regressor Prediction'] = predictions

output_df
```

### IN LINEAR REGRESSION 

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score

tuned_parameters = [{'alpha': [0.1, 1.0, 10.0]}]

clf = GridSearchCV(
    estimator=Ridge(max_iter=5000),
    param_grid=tuned_parameters,
    scoring='neg_mean_squared_error',
    cv=5,
    return_train_score=True,
    verbose=1
)

clf.fit(X_train_transformed, y_train)

# Get the best parameters and best model
best_params = clf.best_params_
best_model = clf.best_estimator_

# Make predictions
predictions = best_model.predict(X_test_transformed)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Best Parameters:", best_params)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

```python
output_df = pd.DataFrame({'Actual': y_test})
output_df['Linear regressor Prediction'] = predictions

output_df
```

### IN DT 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

tuned_parameters = [{'max_depth': [5, 10, 15], 'min_samples_split': [2, 5, 10]}]

clf = GridSearchCV(
    estimator=DecisionTreeRegressor(),
    param_grid=tuned_parameters,
    scoring='neg_mean_squared_error',  # Use appropriate regression scoring metric
    cv=5,
    return_train_score=True,
    verbose=1
)
clf.fit(X_train_transformed, y_train)

# Get the best parameters and best model
best_params = clf.best_params_
best_model = clf.best_estimator_

# Make predictions
predictions = best_model.predict(X_test_transformed)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Best Parameters:", best_params)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

```python
output_df = pd.DataFrame({'Actual': y_test})
output_df['Desicion Tree Predictions'] = predictions

output_df
```

### IN RF

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

tuned_parameters = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10]
}

clf = GridSearchCV(
    estimator=RandomForestRegressor(),
    param_grid=tuned_parameters,
    scoring='neg_mean_squared_error',  # Use appropriate regression scoring metric
    cv=5,
    return_train_score=True,
    verbose=1
)

clf.fit(X_train_transformed, y_train)

# Get the best parameters and best model
best_params = clf.best_params_
best_model = clf.best_estimator_

# Make predictions
predictions = best_model.predict(X_test_transformed)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Best Parameters:", best_params)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

```python
output_df = pd.DataFrame({'Actual': y_test})
output_df['Random Tree Predictions'] = predictions

output_df
```
