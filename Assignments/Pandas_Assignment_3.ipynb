```python
import pandas as pd
```

```python
df = pd.read_csv("attendance_data.csv")
```

```python
df
```

```python
columns = [col.lower().replace("-", "_").replace("participants","").replace("[","").replace("]","").strip() for col in df.columns ]

df.columns = columns
```

```python
df.columns
```

```python
df["date"] = pd.to_datetime(df["date"], format='%m/%d/%Y')
```

```python
df.head()
```

```python
df["Present"] = (df=="Present").sum(axis =1)
df["Absent"] = (df=="Absent").sum(axis=1)
df["Total_students"]= df["Present"] + df["Absent"]
df["Percentage"]= (df["Present"]/df["Total_students"])*100
```

```python
df.head()
```

```python
df.loc[df["date"].dt.year == 2023,["date","Present","Absent","Percentage"]]
```

```python
df1= df.drop(["timestamp","email address","session",'Present', 'Absent', 'Total_students',
       'Percentage'], axis =1)
```

```python
df1
```

```python
reshaped_df = df1.transpose()
```

```python
reshaped_df.head()
```

```python
# Set the first row as column names
reshaped_df.columns = reshaped_df.iloc[0]

# Remove the first row
reshaped_df = reshaped_df[1:]
reshaped_df.head()
```

```python
reshaped_df.columns
```

```python
reshaped_df["Present"] = (reshaped_df=="Present").sum(axis=1)
```

```python
reshaped_df["Absent"] = (reshaped_df =="Absent").sum(axis=1)
```

```python
reshaped_df["Total_days"] = reshaped_df['Present'] + reshaped_df['Absent']
```

```python
reshaped_df["Percentage"] = (reshaped_df["Present"]/reshaped_df["Total_days"])*100
```

```python
reshaped_df.head()
```

```python
reshaped_df = reshaped_df.replace("Present", "P")
reshaped_df = reshaped_df.replace("Absent", "A")
```
