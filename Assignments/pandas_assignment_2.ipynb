**How many columns can be converted to datetime?**

```python
df['order_placed_at'] = pd.to_datetime(df['order_placed_at'], format = "%Y-%m-%d %H:%M:%S")
```

```python
df['order_delivered_at'] = pd.to_datetime(df['order_delivered_at'], format = "%Y-%m-%d %H:%M:%S.%f")
```

```python
df.info()
```

**What is the Date Range of the order placed?**

```python
df["order_placed_at"].min()
```

```python
df["order_placed_at"].max()
```

**How many pizza's were ordered on 2nd March 2023?**

```python
res = df[(df['order_placed_at'].dt.year == 2023) & (df["order_placed_at"].dt.month == 3)]
```

```python
res
```

```python
df.loc[(df['order_placed_at'].dt.year == 2023) & (df["order_placed_at"].dt.month == 3) & 
       (df['order_placed_at'].dt.day == 2),"order_id"].count()
```

**How many pizza's were ordered between 1st March and 5th March 2023?**

```python
df.loc[(df['order_placed_at'].dt.year == 2023) & (df["order_placed_at"].dt.month == 3) &
       (df['order_placed_at'].dt.day >=1) & (df['order_placed_at'].dt.day <=5), "order_id"].count()
```

**How many orders are placed in total on all Sundays?**

```python
df.loc[df['order_placed_at'].dt.day_name() == "Sunday", "order_id"].count()
```

**On which week day have you observed maximum orders?**

**What is the average Delivery Time taken by delivery boys to deliver the Pizza?**

```python
df["delivery_time"] = df["order_delivered_at"] - df["order_placed_at"]
```

```python
df["delivery_time"].mean()
```

**What is the maximum delivery time taken to deliver a pizza?**

```python
df["delivery_time"].max()
```

**What is the order id with the maximum delivery time?**

```python
df.loc[df["delivery_time"]==df["delivery_time"].max(), "order_id"]
```

**How many pizza's got delivered after 30 minutes and what is the total amount for such orders?**

**This is important as Diminos promise 30 minutes delivery. For any pizza delivered after 30 minutes, customer will get a full refund.**

```python
res = df["delivery_time"].dt.total_seconds()/60
```

```python
df.loc[res>30,"order_amount"].sum()
```

```python
df.loc[res>30,"order_amount"].count()
```
